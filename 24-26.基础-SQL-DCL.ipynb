## DCL-管理用户

<img src="./image/DCL-管理用户.jpg" width="70%">

### 1. 查询用户

```python
%%sql
USE mysql;
SELECT * FROM user;
```

### 2. 创建用户

```python
%%sql
CREATE USER 'ROOKIE'@'localhost' IDENTIFIED BY '123456';
USE mysql;
SELECT * FROM user;
```

### 3. 修改用户密码

```python
%%sql
ALTER USER 'ROOKIE'@'localhost' IDENTIFIED BY '12345';
USE mysql;
SELECT * FROM user;
```

### 4. 删除用户

```python
%%sql
DROP USER 'ROOKIE'@'localhost';
USE mysql;
SELECT * FROM user;
```
